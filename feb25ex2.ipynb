```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV

# Dependencies


# Load in csv
battle_df = pd.read_csv("got.csv")
battle_df
```

```python
# Get attacker and defender king data
total_attacker = battle_df["attacker_king"].count()
total_attacker
```

```python
#per_school_averages = school_data_complete_df.groupby(["school_name"]).mean()
per_attacker_king = battle_df.groupby(["attacker_king"]).count()["battle_number"]
per_attacker_king
```

```python
# Get attacker and defender king data
total_defender = battle_df["defender_king"].count()
total_defender
```

```python
per_defender_king = battle_df.groupby(["defender_king"]).count()["battle_number"]
per_defender_king
```

```python
# Get total battle data #interchanging per_defender_king & per_attacker_king can change  the results 
total_battle_data = per_defender_king .add(per_attacker_king,fill_value=0)
total_battle_data
```

```python
# Configure plot and ticks
total_battle_data.plot(kind="bar",facecolor="red")

# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of battles participated")
plt.xlabel("Kings")
plt.xticks(rotation=90)
# Show plot
plt.show()

# Resize plot to display labels
plt.tight_layout()
```
